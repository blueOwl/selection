```python
import os

path = './M1_res/'
file_list = os.listdir(path)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
```

```python
def draw_file(file_name):
    f = open(file_name)
    f.readline()
    d1, d2 = [], []
    for i in f:
        if not i : continue
        line = i.rstrip().split()
        d1.append(float(line[0]))
        d2.append(float(line[1]))
    plt.plot(list(range(len(d1))),d1,"b--",linewidth=0.5) 
    plt.plot(list(range(len(d2))),d2,"r--",linewidth=0.5) 
```

```python
models = set()
for f in file_list:
    models.add(f[:14])
```

```python
models = list(models)
models
```

```python
def draw_model(model):
    plt.figure(figsize=(8,4))
    for f in file_list:
        if f[:14] == model:
            draw_file(path + f)
            
```

```python
def draw_ratio(file_name):
    f = open(file_name)
    f.readline()
    d = []
    for i in f:
        if not i : continue
        line = i.rstrip().split()
        d.append(float(line[0]) / (sum(map(float, line))))
        
    plt.plot(list(range(len(d))),d,"b--",linewidth=0.5) 
    
draw_ratio(path + file_list[0])
```

```python
def draw_model_ratio(model):
    plt.figure(figsize=(8,4))
    for f in file_list:
        if f[:14] == model:
            draw_ratio(path + f)
            
```
