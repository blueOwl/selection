```python
def beta_08(T, tp):
    #if T < 0: T = -1 * T
    tp = TYPE_MAP[tp]
    beta_l = {	'vert':1, 
		        'horz':0.8}
    #tp is agent type, 0 is vertical, 1 is horizontal
    l = beta_l[tp]
    return l * ((2 * (T - T_MIN)**ALPHA *(T_OPT-T_MIN)** ALPHA) - (T - T_MIN) ** (2 * ALPHA)) / ((T_OPT - T_MIN)**(2 * ALPHA))

def beta_death(T):
    #tp is agent type, 0 is vertical, 1 is horizontal
    l = 1
    return l * ((2 * (T - T_MIN)**ALPHA *(T_OPT-T_MIN)** ALPHA) - (T - T_MIN) ** (2 * ALPHA)) / ((T_OPT - T_MIN)**(2 * ALPHA))
```

```python
class SimpleAgent(GenAgent):
    #def set_beta(self):
    #    self.beta = beta_08
    #    self.beta_death = beta_death
    def check_env_press(self):
        coin = np.random.random()
        if self.gen_type == 1:
            theta = 0.2
        else:
            theta = 0.3
        if coin <= theta:
            self.lifetime = 0
        else:
            self.lifetime = 2    
    def get_gener_p(self):
        return 0.5
    
class SimpleModel(GenModel):
    def get_agent_config(self):
        return SimpleAgent
```

```python
fixed_params = {'N':2000, 'width':100, 'height':100}

variable_params = {'init_ratio': [0.3, 0.4, 0.5, 0.6, 0.7]}

batch_run = BatchRunnerMP(
    SimpleModel, 5,
    variable_parameters=variable_params,
    fixed_parameters=fixed_params,
    iterations=2,
    max_steps=50,
    model_reporters={"two type ratio": compute_ratio}
)
batch_run.run_all()
run_data = batch_run.get_model_vars_dataframe()
```
